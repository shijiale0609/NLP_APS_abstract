```
#!/usr/bin/python
# -*- coding: UTF-8 -*-
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
import re
```

```
url = "https://meetings.aps.org/Meeting/MAR21/SessionIndex2"
html = urlopen(url).read().decode('utf-8')
soup = BeautifulSoup(html, features="html.parser")
```

```
sessions = soup.find_all('a',href=re.compile(r"/Meeting/MAR21/Session"))
```

```
for i in range(0, len(sessions)):
  print ("https://meetings.aps.org" + sessions[i]['href'])
```

```
f = open('sessions_link.txt', 'a+')
for i in range(0, len(sessions)):
  f.write("https://meetings.aps.org" + sessions[i]['href'])
  f.write('\n')
f.close()
```

```
f = open('sessions_topics.txt', 'a+')
for i in range(0, len(sessions)):
  f.write(sessions[i].get_text())
  f.write('\n')
f.close()
```

```
sessions_link = []
with open("sessions_link.txt") as f:
  for line in f:
    cols = line.split()
    #print (cols)
    sessions_link.append(cols[0])
```

```
talk_links = []
for i in range(0, len(sessions_link)):
  if i%50 ==0: print (i)
  url = sessions_link[i]
  html = urlopen(url).read().decode('utf-8')
  soup = BeautifulSoup(html, features="html.parser")
  sessions = soup.find_all('a',href=re.compile(r"/Meeting/MAR21/Session"))
  for i in range(0, len(sessions)):
     talk_link = "https://meetings.aps.org" + sessions[i]['href']
     talk_links.append(talk_link)
```

```
List = []
for i in range(0,len(talk_links)):
  if not talk_links[i] in List:
    List.append(talk_links[i]) 

print (len(List))
```

```
f1 = open('talk_links.txt', 'a+')
for i in range(0, len(List)):
  if len(List[i])== 52 or len(List[i])==53:
    f1.write(List[i])
    f1.write('\n')
f1.close()
```
